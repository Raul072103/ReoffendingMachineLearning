```python
imp_constant = SimpleImputer(strategy='constant', fill_value=0)
imp_mean = SimpleImputer(missing_values = np.nan, strategy='mean')
```

```python
#I make a copy in case i screw up something
copy_df = df.copy()

#First i want to change only the columns where the 2 values are both null and replace them with constants
drug_tests = ["days_between_drug_tests", "marijuana_positive_tests", "cocaine_positive_tests", "meth_positive_tests", "other_positive_tests"]
copy2_df = copy_df[drug_tests] 
condition = copy2_df["days_between_drug_tests"].isnull() & copy2_df["marijuana_positive_tests"].isnull() & copy2_df["cocaine_positive_tests"].isnull() & copy2_df["meth_positive_tests"].isnull() & copy2_df["other_positive_tests"].isnull() 
copy2_df.loc[condition] = imp_constant.fit_transform(copy2_df[condition])

#Now I place the changes in my copy
copy_df[drug_tests] = copy2_df

#Second let's change those values where there is no record for "days_between_drug_tests" but from the ratio of positive tests we can deduce that tests were made
drug_tests = ["days_between_drug_tests"]
copy2_df = copy_df[drug_tests] 
copy2_df = imp_mean.fit_transform(copy2_df)

#Now I place the changes in my copy
copy_df[drug_tests] = copy2_df

#Place in the main df the changes I have made
df = copy_df.copy()
```

Now that I have dealt with those features I still have some features that have NaN values and I need to impute them: risk_score, days_employed and jobs_per_year.

* **risk_score**: As I said above there are cases where no risk_score is given to an inmate, so after I do a second check and I look at the distribution of felonies and more, I will decide if i will replace all NaN with 1 (since risk_score is from 1 to 10, I think giving the 0 is not appropiate)
* **days_employed** and **jobs_per_year**: Here I think it is the best to replace the NaN by the mean of the others

```python
#This is to see the distribution of fellonies and more where risk_score is null
copy2_df = df[df["risk_score"].isnull()]

for ft in nominal_features:
    unique_ft = copy2_df[ft].unique()
    unique_dictionary = {}
    for ft2 in unique_ft:
        unique_dictionary[ft2] = 0
        for i in copy2_df[ft]:
            if i == ft2:
                unique_dictionary[ft2] += 1

    print(ft, unique_dictionary)
```
