## Data splitting and data visualitzation

Before splitting the data I would like to take a look at the dataset and the distribution of the "reoffended_within_3_years" and also the fact that there are aprox 3000 females out of 25000 so I would like to split the data so I don't end up with an uneven distribution of classes.

```python
# I cannot assess the distribution of numeric_features therefore i will assess the distribution of nominal + boolean features

total_features = nominal_features + boolean_features + ["reoffended_within_3_years"]

copy_df = df.copy()
ft_size = 25835 #how many columns there are  
max = 0
name = ""

for ft in total_features:
    dict = {}
    unique_ft = copy_df[ft].unique()
    #I am not particularly interested in their numbers more than I am in their distribution ratio which will be (number of ft in column) / (total nr of ft in column)
    for cat in unique_ft:
        dict[cat] = 0
        size_tuple = copy_df[copy_df[ft] == cat].shape #it is easier like this to get the sizes
        dict[cat] = size_tuple[0] / ft_size # now we calculate the ratio of distribution
        if dict[cat] > max:
            max = dict[cat]
            name = ft
        
    print(ft, dict)
```

There are features which are aproximately evenly distributed, and some which aren't. I am pretty sure I cannot divide the dataset into dev set and test set such that every feature is evenly distributed. However, what I would like to try is to split the dataset at least by taking into account `gender, gang_affiliated and race`(even though the distribution of the races seems ok (0.42 - 0.57), this is a sensible subject, so I should try to ensure evenly distribution).
